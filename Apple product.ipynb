```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('apple_products.csv')

df
```

```python

# Data type conversion
df['Sale Price'] = pd.to_numeric(df['Sale Price'])

df['Discount Percentage'] = pd.to_numeric(df['Discount Percentage'])
df['Number Of Ratings'] = pd.to_numeric(df['Number Of Ratings'])
df['Number Of Reviews'] = pd.to_numeric(df['Number Of Reviews'])
df['Star Rating'] = pd.to_numeric(df['Star Rating'])

# Feature Engineering
df['Price Difference'] = df['Mrp'] - df['Sale Price']
df['Discounted Price Ratio'] = df['Sale Price'] / df['MRP']
df['Popularity Metric'] = df['Number Of Ratings'] + df['Number Of Reviews']

# Outlier detection
df = df[(df['Sale Price'] > 0) & (df['Sale Price'] < df['MRP'] * 2)]
```
